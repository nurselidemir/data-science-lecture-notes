```python
yasListesi = [10, 20, 30, 30, 30, 40, 50, 60, 70, 75]
```

```python
kiloListesi = [20, 60, 80, 85, 86, 87, 70, 90, 95, 90]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi 
```

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "r" )
plt.xlabel("yaş")
plt.ylabel("kilo")
plt.title("Kilonun yaşa göre değişimi")
```

```python
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1 
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

numpyDizis1^3 demektir

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g")
```

```python
plt.show()
```

```python
plt.plot(numpyDizisi1, numpyDizisi2, "g*-")
```

```python
plt.show()
```

```python
plt.subplot(1,2,1)    # 1 sıra olacak, 2 kolon olacak suan 1.grafik çizilecek
plt.plot(numpyDizisi1, numpyDizisi2, "r*-")
plt.subplot(1,2,2)    # 1 sıra olacak, 2 kolon olacak suan 2.grafik çizilecek
plt.plot(numpyDizisi2, numpyDizisi2, "b--")
```

```python
plt.show()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.3,0.3])
figureAxes.plot(numpyDizisi1, numpyDizisi2, "g")
figureAxes.set_xlabel("x ekseni verileri")
figureAxes.set_ylabel("y ekseni verileri")
figureAxes.set_title("Grafik Başlığı")
```

```python
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.22,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1, numpyDizisi2, "g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2, numpyDizisi1, "b")
eksen2.set_xlabel("x ekseni küçük")
eksen2.set_ylabel("y ekseni küçük")
eksen2.set_title("Küçük Grafik Başlığı")

plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows = 1, ncols = 2)
type(benimEksen)

for eksen in benimEksenler : 
    eksen.plot(numpyDizisi1, numpyDizisi2, "g")
    eksen.set_xlabel("x ekseni")
plt.show()
```

```python
yeniFigure = plt.figure()
# parantez içine dpi = 150 kaydetmeden önce belirtmek daha mantıklı

yeniEksen = yeniFigure.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 2, label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3 , label = "numpyDizisi ** 3")
yeniEksen.legend(loc = 0)

plt.show()
```

```python
yeniFigure.savefig("benimfigure.png", dpi = 200)
```
